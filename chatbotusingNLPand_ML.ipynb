```python
import pandas as pd
import random
import nltk

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

nltk.download('punkt')
```

```python
data = {
    "text":[
        "hello",
        "hi",
        "hey",
        "good morning",
        "good evening",

        "bye",
        "goodbye",
        "see you later",

        "thanks",
        "thank you",
        "thanks a lot",

        "what is ai",
        "what is machine learning",
        "explain deep learning",

        "how are you",
        "are you fine",
        "how is your day"
    ],

    "intent":[
        "greeting",
        "greeting",
        "greeting",
        "greeting",
        "greeting",

        "goodbye",
        "goodbye",
        "goodbye",

        "thanks",
        "thanks",
        "thanks",

        "ai_question",
        "ai_question",
        "ai_question",

        "status",
        "status",
        "status"
    ]
}

df = pd.DataFrame(data)

print("Dataset Loaded")
print(df.head())
```

```python
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(df["text"])
y = df["intent"]
```

```python
model = LogisticRegression()
model.fit(X, y)

print("Model Trained ‚úÖ")
```

```python

responses = {
    "greeting":[
        "Hello üëã",
        "Hi there üòä",
        "Hey! How can I help?"
    ],

    "goodbye":[
        "Goodbye üëã",
        "See you later",
        "Take care"
    ],

    "thanks":[
        "You are welcome üòä",
        "No problem üëç",
        "Happy to help"
    ],

    "ai_question":[
        "AI means machines that mimic human intelligence.",
        "Machine Learning is a subset of AI.",
        "Deep Learning uses neural networks."
    ],

    "status":[
        "I am working perfectly üòä",
        "I am good, thanks for asking",
        "All systems running fine üöÄ"
    ]
}
```

```python
def chatbot_reply(user_text):

    text_vec = vectorizer.transform([user_text])
    intent = model.predict(text_vec)[0]

    return random.choice(responses[intent])
```

```python
print("\nü§ñ ML Chatbot Ready (Type 'quit' to exit)\n")
```

```python
while True:

    user_input = input("You: ")

    if user_input.lower() == "quit":
        print("Bot: Goodbye üëã")
        break

    reply = chatbot_reply(user_input)
    print("Bot:", reply)
```
